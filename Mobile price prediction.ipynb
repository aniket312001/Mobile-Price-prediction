```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scale_feature = scaler.fit_transform(df.drop('price_range',axis=1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('price_range',axis=1),df['price_range'], test_size=0.33, random_state=42)
# X_train, X_test, y_train, y_test = train_test_split(scale_feature,df['price_range'], test_size=0.33, random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

lg = LogisticRegression()
lg.fit(X_train,y_train)
lg_predict = lg.predict(X_test)

print(classification_report(y_test,lg_predict))
print('\n')
print(confusion_matrix(y_test,lg_predict))
print('\n')
print('Accuracy -> ',accuracy_score(y_test,lg_predict))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_predict = rf.predict(X_test)

print(classification_report(y_test,rf_predict))
print('\n')
print(confusion_matrix(y_test,rf_predict))
print('\n')
print('Accuracy -> ',accuracy_score(y_test,rf_predict))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train,y_train)
knn_predict = knn.predict(X_test)

print(classification_report(y_test,knn_predict))
print('\n')
print(confusion_matrix(y_test,knn_predict))
print('\n')
print(accuracy_score(y_test,knn_predict))
```

```python
# Elbow Method For optimum value of K
error_rate =[]

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,'y.-')
plt.show
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train,y_train)
knn_predict = knn.predict(X_test)

print(classification_report(y_test,knn_predict))
print('\n')
print(confusion_matrix(y_test,knn_predict))
print('\n')
print('Accuracy ->',accuracy_score(y_test,knn_predict))
```
